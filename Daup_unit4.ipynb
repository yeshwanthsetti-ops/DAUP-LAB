```python
#Write a Pandas program to create and display a one-dimensional array-like object containing an array of data using Pandas module.
import pandas as pd

data = [18, 45, 7, 33, 17, 333]
series = pd.Series(data)

print(series)
```

```python
#Write a Pandas program to convert a Pandas Series to Python list and its types.
import pandas as pd

data = [19, 87, 88, 61, 6, 15]
series = pd.Series(data)

list_data = series.tolist()
print(list_data)
print(type(list_data))
```

```python
#Write a Pandas program to create and display a DataFrame from a specified dictionary data which has the index labels.
import pandas as pd
import numpy as np

exam_data = {
    'name': ['Kailash', 'Ram', 'Raju', 'Mani', 'Emily', 'Anil', 'Sai', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)

print(df)
```

```python
#Write a Pandas program to change the name 'James' to 'Suresh' in name column of the DataFrame.
# Replace 'Kailash' with 'Ajay' in the 'name' column
df.replace({'name': {'Kailash': 'Bhadri'}}, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
#Write a Pandas program to insert a new column in existing DataFrame.
# Add a new column 'grade' to the DataFrame
df['grade'] = ['A','B','A','C','B','A','B','C','C','A']

# Display the updated DataFrame
print(df)
```

```python
#Write a Pandas program to get list from DataFrame column headers.
# Convert DataFrame column names to a list
columns_list = list(df.columns)

# Display the list of columns
print(columns_list)
```
