```python
#Write a Python program to calculate the nth Fibonacci number using a function.
def fibonacci(num):
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fibonacci(num - 1) + fibonacci(num - 2)

print("Fibonacci of 5:", fibonacci(6))
```
